## Markdown文件格式解释

Markdown 是一种轻量级标记语言，它的设计理念就是让文档看起来就像纯文本一样简洁易读，同时又能方便地转换成富文本格式。下面我将对常用的 Markdown 符号和语法进行解释。

## 1. 标题

使用 `#` 符号来表示标题，数量从 1 到 6，对应 HTML 中的 `<h1>` 到 `<h6>`。

```markdown
# 这是一级标题
## 这是二级标题
### 这是三级标题
```
## 2. 强调
###### 使用 * 或 _ 来表示斜体和粗体。
*这是斜体文本*
**这是粗体文本**
***这是粗斜体文本***

## 3. 列表
### 3.1 无序列表
###### 使用 *、+ 或 - 作为列表标记。
* 项目一
* 项目二
  * 子项目（通过缩进）
### 3.2 有序列表
###### 使用数字接着一个英文句点。
1. 第一项
2. 第二项
3. 第三项
## 4. 链接
###### 链接文本放在 [ ] 中，链接地址放在随后的 ( ) 中。
[这是一个链接](https://www.example.com)
[这是一个带标题的链接](https://www.example.com "示例网站")
## 5. 图片
###### 图片的语法与链接类似，但在前面多一个 !
###### 图片替代文本：如果图片无法显示，将显示这段文字。
###### 可选标题：鼠标悬停在图片上时显示的文字。
![图片替代文本](图片的URL "可选标题")
## 6. 引用
###### 使用 > 符号来表示引用。
> 这是一个引用块。
> 引用块可以包含多行。
## 7. 代码
### 7.1 行内代码
###### 使用反引号 ` 将代码包裹起来。
使用 `printf()` 函数来打印输出。
### 7.2 代码块
###### 使用三个反引号 ``` 来包裹一段代码。你还可以在开始的反引号后指定语言名称，以实现语法高亮。
```python
def hello_world():
    print("Hello, World!")
```
## 8. 表格
###### 使用连字符 - 来分隔表头和内容，使用管道符 | 来分隔列。
| 左对齐 | 居中对齐 | 右对齐 |
| :----- | :------: | -----: |
| 单元格 |  单元格  | 单元格 |
| 单元格 |  单元格  | 单元格 |

## 9. 分割线
###### 使用三个或更多的 *、- 或 _ 来创建一条水平分割线。
***
---
___
## 10. 转义字符
###### 如果你想要显示原本用于 Markdown 格式的字符（如 *、_、# 等），你需要在它们前面加上反斜杠 \。
\*这不会变成斜体\*
## 11. 删除线
###### 使用两个波浪号 ~~ 包裹文本来表示删除线。
~~这段文字被删除了。~~
## 12. 任务列表
###### 使用 - [ ] 表示未完成的任务，使用 - [x] 表示已完成的任务。
- [x] 已完成的任务
- [ ] 未完成的任务
## 13. 数学公式
### 13.1 行内公式
###### 使用 $ 符号包裹公式。
勾股定理：$a^2 + b^2 = c^2$
### 13.2 块级公式
###### 使用 $$ 符号包裹公式。
$$
a^2 + b^2 = c^2
$$
###### 提示： 以上是 Markdown 最核心和最常用的语法。掌握了它们，你就能写出结构清晰、格式美观的文档了。
